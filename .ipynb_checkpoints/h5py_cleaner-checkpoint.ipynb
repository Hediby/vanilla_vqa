```python
import h5py
import numpy as np
import json
```

```python
for split in "train val".split():
    print(split)
    bb_dict = {}
    for l in open('datasets/vqa/%s/edgeboxes.json' % split,'r'):
        bb_dict.update(json.loads(l))
    vocabulary = {}
    for i,l in enumerate(open('datasets/vqa/vocabulary.txt','r')):
        vocabulary[l.strip()] = i
    ground_truth = []
    multiple_choices = []
    images = []
    questions = []
    maxlen_gt, maxlen_mc, maxlen_q = 0,0,0
    dataset = open('datasets/vqa/%s/dataset_idxs.json' % split,'r')
    BBS = []
    img_list = []
    for k in bb_dict:
        v = bb_dict[k]
        img_list.append(int(k))
        b = bb_dict[k]
        bb = [b[j % len(b)] for j in range(100)]
        BBS.append(bb)
    BBS = np.array(BBS)
    for l in dataset:
        l = json.loads(l)
        maxlen_gt = max(maxlen_gt,
                        len(max(l['ground_truth'], 
                                key=len)))
        maxlen_mc = max(maxlen_mc,
                        len(max(l['multiple_choices'], 
                                key=len)))
        maxlen_q = max(maxlen_q,len(l['question']))
        questions.append(l['question'])
        ground_truth.append(l['ground_truth'])
        multiple_choices.append(l['multiple_choices'])
        images.append(l['image_id'])
    print(maxlen_gt, maxlen_mc, maxlen_q)
    dataset.close()
    BBS = np.array(BBS)
    N = len(multiple_choices)
    MC = np.zeros((N,18,maxlen_mc),dtype='int') + vocabulary['</s>']
    GT = np.zeros((N,10,maxlen_gt),dtype = 'int') + vocabulary['</s>']
    Q = np.zeros((N,maxlen_q),dtype = 'int') + vocabulary["</s>"]
    for i in range(N):
        for j in range(MC.shape[1]):
            for k in range(len(multiple_choices[i][j])):
                MC[i,j,k] = multiple_choices[i][j][k]

        for j in range(GT.shape[1]):
            for k in range(len(ground_truth[i][j])):
                GT[i,j,k] = ground_truth[i][j][k]

        for j in range(len(questions[i])):
            Q[i,j] = questions[i][j]
            
    f = h5py.File('/home/hbenyounes/vqa/datasets/vqa/%s/dataset_cleaner.h5' % split)
    f.create_dataset("ground_truth",dtype="int",data=GT)
    f.create_dataset('image_ids',dtype="int",data=images)
    f.create_dataset("questions",data=Q)
    f.create_dataset("multiple_choices",dtype="int",data=MC)
    f.create_dataset("img_list",data=list(int(x) for x in bb_dict.keys()))
    f.create_dataset("bounding_boxes",data=BBS)
    f.close()
```
